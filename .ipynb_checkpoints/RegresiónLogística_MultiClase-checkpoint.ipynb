```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import os
from sklearn.linear_model import LogisticRegression
```

```python
path = "/Users/USER/Desktop/Universidad/X/Inteligencia Artificial/II Parcial/DataSets"
filename = "iris.csv"
fullpath = os.path.join(path, filename)
```

```python
data= pd.read_csv(fullpath)
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Conversión de datos categoricos a númericos
le = LabelEncoder()
data['variety'] = le.fit_transform(data['variety'])
data.head()
```

```python
data['variety'].unique()
```

* Setosa = 0
* Versicolor = 1
* Virgínica = 2

```python
data.tail()
```

```python
X = data[['sepal.length', 'sepal.width','petal.length','petal.width']].values
Y =data['variety'].values
print(X.shape, Y.shape)
```

```python
# lbfgs para problemas multiclase en datsets pequeños
lo = LogisticRegression(solver='lbfgs')
lo = lo.fit(X , Y)
Y_pred = lo.predict(X)
```

```python
lista = list(zip(Y_pred, Y))
lista_df = pd.DataFrame(lista, columns = ['Y_pred', 'Y'])
lista_df.head(20)
```

```python
pd.crosstab(Y_pred,Y)
```
